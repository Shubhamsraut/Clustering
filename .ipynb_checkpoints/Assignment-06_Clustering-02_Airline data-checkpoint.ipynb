```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sn
```

```python
airline1 = pd.read_csv("EastWestAirlines.csv")
airline1
```

### 1. EDA

```python
airline1.info()
```

```python
airline2 = airline1.drop(['ID#'],axis=1)
airline2
```

### 1. Hierarchical Clustering

```python
# Standardize the data:
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return(x)
```

```python
# (considering the numerical part of data)
airline2_norm = norm_func(airline2)
airline2_norm
```

```python
# creating a dendrogram
dendrogram = sch.dendrogram(sch.linkage(airline2_norm ,method ='average'))
```

```python
# create clusters
hc =  AgglomerativeClustering(n_clusters = 4, affinity = 'euclidean', linkage = 'average')
hc
```

```python
# clusters for chart
y_hc =hc.fit_predict(airline2_norm)
y_hc
```

```python
Clusters = pd.DataFrame(y_hc, columns = ['Clusters'])
Clusters
```

```python
airline2_norm['h_clusterid'] = y_hc
```

```python
airline2_norm
```

### 2. K-means

```python
# How to find optimum number of  cluster
#The K-means algorithm aims to choose centroids that minimise the inertia, or within-cluster sum-of-squares criterion:
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
```

```python
wcss = []
for r in range (1,11):
    kmeans = KMeans(n_clusters = r, random_state = 0)
    kmeans.fit(airline2_norm)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1,11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Build Cluster algorithm
from sklearn.cluster import KMeans
clusters_new =  KMeans (5, random_state = 42) # Random state --> checking that your results are stable across a number of different distinct random seeds. Popular integer random seeds are 0 and 42.
clusters_new.fit(airline2_norm)
```

```python
clusters_new.labels_
```

```python
# Assign clusters to the data set
airline2_norm['kmeans_clusterid_new'] = clusters_new.labels_
```

```python
#These are standardized values.
clusters_new.cluster_centers_
```

```python
airline2_norm.groupby('kmeans_clusterid_new').agg(['mean']).reset_index()
```

### 3. DBSCAN

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
airline2.info()
```

```python
airline3 = airline2.iloc[:,1:11]
airline3
```

```python
airline3.values
```

```python
stscaler = StandardScaler().fit(airline3.values)
x = stscaler.transform(airline3.values)
```

```python
x
```

```python
dbscan = DBSCAN(eps=2,min_samples=4)
dbscan.fit(x)
```

```python
#Noisy samples given the label--> -1.
dbscan.labels_
```

```python
cl =pd.DataFrame(dbscan.labels_,columns=['DBSCAN_cluster'])
```

```python
cl
```

```python
pd.concat([airline3,cl],axis=1)
```
